```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = dataset.drop("Unnamed: 0", axis=1)
print(dataset.head(5))
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=dataset, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
gb=dataset.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
plt.title('parent education impact on student score')
sns.heatmap(gb,annot=True)
plt.show()
```

```python
gbx=dataset.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gbx)
```

```python
plt.figure(figsize=(4,4))
plt.title('parent marital status impact on student score')
sns.heatmap(gbx,annot=True)
plt.show()
```

```python
groupA=dataset.loc[(dataset['EthnicGroup'] == 'group A')].count()
groupB=dataset.loc[(dataset['EthnicGroup'] == 'group B')].count()
groupC=dataset.loc[(dataset['EthnicGroup'] == 'group C')].count()
groupD=dataset.loc[(dataset['EthnicGroup'] == 'group D')].count()
groupE=dataset.loc[(dataset['EthnicGroup'] == 'group E')].count()
plt.figure(figsize=(4,4))
l=["groupA","groupB","groupC","groupD","groupE"]
mlist=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Group")
plt.show()
```
